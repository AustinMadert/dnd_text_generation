```python
def create_monster_txt_spec_char(df):
    monsters_list = []
    for row in range(df.shape[0]):
        monster = '<<start>>'
        for idx, col in enumerate(df.columns):
            monster += ("[[" + col + "]]" + str(df.iloc[row, idx]))
        monster +='<<end>>'
        monsters_list.append(monster)
    return ''.join(monsters_list), monsters_list

def create_monster_txt(df):
    monsters_list = []
    for row in range(df.shape[0]):
        monster = '<<start>>'
        for idx, col in enumerate(df.columns):
            monster += (col + ":" + str(df.iloc[row, idx]))
        monster +='<<end>>'
        monsters_list.append(monster)
    return ''.join(monsters_list), monsters_list
```

```python
monsters_string, monsters_list = create_monster_txt_spec_char(adjdf_with_full_abilities)
```

```python
with open('adjusted_full_abilities.txt', 'w') as f:
    f.write(monsters_string)
```
