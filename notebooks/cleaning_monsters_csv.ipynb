```python
def paren_remove(text):
    return text.replace('(', '').replace(')', '').strip()[:2]
```

```python
df['charisma'] = df.charisma.apply(paren_remove)
df['dexterity'] = df.dexterity.apply(paren_remove)
df['strength'] = df.strength.apply(paren_remove)
df['intelligence'] = df.intelligence.apply(paren_remove)
df['wisdom'] = df.wisdom.apply(paren_remove)
df['constitution'] = df.constitution.apply(paren_remove)
```

```python
def paren_remove_hp(text):
    return text.replace('(', '').replace(')', '').strip()[:3]
```

```python
df['hit_points'] = df.hit_points.apply(paren_remove_hp)
```

```python
def hp_fix(text):
    result = re.search(' ', text)
    if result:
        return(df.iloc[13,6][:2])
    else:
        return text
```

```python
df['hit_points'] = df.hit_points.apply(hp_fix)
```

```python
def create_monster_txt(df):
    monsters_list = []
    for row in range(df.shape[0]):
        monster = '<<start>>'
        for col in df.columns:
            monster += ("[[" + col + "]]" + str(df.loc[row, col]))
        monster +='<<end>>'
        monsters_list.append(monster)
    return ''.join(monsters_list), monsters_list
```

```python
newdf = df.drop(['url', 'challenge', 'skills', 'languages'], axis=1)
```

```python
def stripper(text):
    return text.strip().lower()
```

```python
newdf = newdf.applymap(stripper)
```

```python
colist = ['strength', 'dexterity', 'constitution', 'wisdom', 'intelligence', 'charisma', 
          'armor_class', 'hit_points', 'speed', 'actions', 'short_desc', 'name']
```

```python
newdf = newdf[colist]
```
